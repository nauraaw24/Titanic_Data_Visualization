```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load dataset Titanic dari seaborn
df = sns.load_dataset("titanic")

# Menampilkan beberapa baris pertama
print(df.head())
```

```python
# Visualisasi jumlah penumpang selamat vs tidak selamat
plt.figure(figsize=(6,4))
sns.countplot(data=df, x="survived", palette="pastel")
plt.xticks([0, 1], ["Tidak Selamat", "Selamat"])
plt.xlabel("Status Keselamatan")
plt.ylabel("Jumlah Penumpang")
plt.title("Jumlah Penumpang yang Selamat dan Tidak Selamat")
plt.show()
```

```python
# Visualisasi hubungan kelas tiket dan keselamatan
plt.figure(figsize=(6,4))
sns.barplot(data=df, x="pclass", y="survived", ci=None, palette="muted")
plt.xticks([0, 1, 2], ["Kelas 1", "Kelas 2", "Kelas 3"])
plt.xlabel("Kelas Tiket")
plt.ylabel("Rata-rata Keselamatan")
plt.title("Pengaruh Kelas Tiket terhadap Keselamatan")
plt.show()
```

```python
# Visualisasi distribusi usia penumpang
plt.figure(figsize=(8,5))
sns.histplot(df["age"].dropna(), bins=20, kde=True, color="skyblue")
plt.xlabel("Usia")
plt.ylabel("Frekuensi")
plt.title("Distribusi Usia Penumpang Titanic")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.select_dtypes(include=['number']).corr(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Heatmap Korelasi Variabel Titanic")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x="pclass", y="age", palette="Set2")
plt.xticks([0, 1, 2], ["Kelas 1", "Kelas 2", "Kelas 3"])
plt.xlabel("Kelas Tiket")
plt.ylabel("Usia Penumpang")
plt.title("Distribusi Usia Berdasarkan Kelas Tiket")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.swarmplot(data=df, x="sex", y="age", hue="survived", palette="coolwarm", dodge=True)
plt.xlabel("Jenis Kelamin")
plt.ylabel("Usia Penumpang")
plt.title("Perbandingan Jenis Kelamin dan Keselamatan")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.kdeplot(df["fare"], shade=True, color="green")
plt.xlabel("Harga Tiket")
plt.ylabel("Density")
plt.title("Distribusi Harga Tiket Titanic")
plt.show()
```
